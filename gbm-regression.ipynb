```python
import h2o
```

```python
h2o.init()
```

```python
from h2o.estimators.gbm import H2OGradientBoostingEstimator
```

```python
data = h2o.import_file("/kaggle/input/real-estate-dataset/data.csv")
```

```python
data.head()
```

# Regression Model

##### *We don't define response variable as factor for Regression problems (continuous values of response)*

```python
y = "MEDV" # response
```

```python
preds = data.col_names
preds.remove(y) # predictors
```

# Split into Training and Testing Datasets

```python
train, test = data.split_frame(ratios = [0.75]) # training = 75%, testing = 25%
```

# GBM

```python
gbm_default = H2OGradientBoostingEstimator(model_id = "gbm_default", seed = 1234)
```

```python
gbm_default.train(x = preds, y = y, training_frame = train)
```

# Check Model Performance on Testing Dataset

```python
test_gbm = gbm_default.predict(test)
```

```python
test_gbm.head(5)
```

```python
perf = gbm_default.model_performance(test)
perf
```

```python
gbm_default.r2()
```

# Decide on GBM Regression Parameters with GridSearch

```python
grid_search_gbm = H2OGradientBoostingEstimator(
stopping_rounds = 25,
stopping_metric = "rmse",
col_sample_rate = 0.65,
sample_rate = 0.65,
seed = 1234
)
```

```python
from h2o.grid.grid_search import H2OGridSearch
```

```python
hyper_params = {
    'learn_rate': [0.01, 0.02, 0.03],
    'max_depth': [4, 8, 16, 24],
    'ntrees': [50, 250, 1000]
}

grid = H2OGridSearch(grid_search_gbm, hyper_params,
                    grid_id = 'depth_grid',
                    search_criteria = {'strategy': 'Cartesian'})

# Train Grid Search
grid.train(x = preds, y = y, training_frame = train)
```

```python
gbm_gridperf1 = grid.get_grid(sort_by = 'rmse', decreasing = True)
gbm_gridperf1
```

```python
best_gbm1 = gbm_gridperf1.models[-1]
```

```python
best_gbm1.rmse()
```

```python
best_gbm_perf1 = best_gbm1.model_performance(test)
```

```python
best_gbm_perf1
```

```python
best_gbm_perf1.rmse()
```
